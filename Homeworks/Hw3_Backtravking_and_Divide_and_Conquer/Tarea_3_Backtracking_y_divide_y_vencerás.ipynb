## Actividad 1. Implementación del Problema de la Mochila con Backtracking

Descripción del Problema:

El Problema de la Mochila consiste en lo siguiente: Dado un conjunto de objetos, cada uno con un peso y un valor, determina la cantidad máxima de valor que se puede obtener al llenar una mochila de capacidad limitada. Los estudiantes deben implementar una solución eficiente para este problema utilizando el enfoque de backtracking.

Instrucciones:

Implementa una función en Python que resuelva el Problema de la Mochila utilizando el enfoque de backtracking.

La función debe tomar como entrada:

- Un arreglo de objetos, donde cada objeto se representa como una tupla (peso, valor).
- La capacidad máxima de la mochila (un número entero positivo).
- La función debe devolver la máxima cantidad de valor que se puede obtener al colocar la mayor cantidad de objetos en la mochila sin exceder su capacidad.

Asegúrate de manejar correctamente casos en los que haya objetos con pesos mayores que la capacidad de la mochila.

Implementa una función adicional que genere un conjunto de objetos aleatorios para probar tu implementación. Puedes utilizar bibliotecas de generación de números aleatorios si es necesario.



```python
from itertools import combinations
import random

def getPeso(mochila):
    peso = 0
    for caja in mochila:
        peso += caja[0]
    return peso

def getCosto(mochila):
    costo = 0
    for caja in mochila:
        costo += caja[1]
    return costo

def verificarSolucion(mochila, capacidad):
    peso = 0
    for caja in mochila:
        peso += caja[0]

    return peso <= capacidad


def obtenerMejor(lista):
    costoMayor = 0
    for elemento in lista:
        costoA = getCosto(elemento)
        if costoA > costoMayor:
            costoMayor = costoA
    return costoMayor

def filtrarSoluciones(lista, capacidad):
    solValidas = []
    for i in lista:
        if verificarSolucion(i, capacidad):
            solValidas.append(i)

    return solValidas


def evaluarPaquetes(lista, capacidad):
    soluciones = []
    for i in range(1, len(lista) + 1):
        combi = combinations(lista, i)
        listaDeep = [list(combinacion) for combinacion in combi]

        soluciones.extend(listaDeep)

    nueva = filtrarSoluciones(soluciones, capacidad)
    costoM = obtenerMejor(nueva)
    return costoM

def generarPaquetes(numPaquetes):
    packs = []
    for i in range(numPaquetes):
        peso = random.randint(0, 50)
        costo = random.randint(0, 150)
        packs.append((peso, costo))
    return packs

#(peso, valor)
numPaquetes = random.randint(1, 10)
paquetes = generarPaquetes(numPaquetes)
print("Paquetes: ", paquetes)
#paquetes = [(10, 60), (20, 100), (30, 120), (40, 100)]
capacidadMaxima = random.randint(0, 200)
print("Capacidad Maxima: ", capacidadMaxima)
costM = evaluarPaquetes(paquetes, capacidadMaxima)
print("Solucion: ", costM)
```

```python
def Solution(arr1,arr2):
    arr = arr1+arr2
    arr.sort()
    n = len(arr)

    # If length of array is even
    if n % 2 == 0:
        z = n // 2
        e = arr[z]
        q = arr[z - 1]
        ans = (e + q) / 2
        return ans

    # If length of array is odd
    else:
        z = n // 2
        ans = arr[z]
        return ans
```

```python
def areDistinct(str, i, j):

    # Note : Default values in visited are false
    visited = [0] * (256)

    for k in range(i, j + 1):
        if (visited[ord(str[k])] == True):
            return False

        visited[ord(str[k])] = True

    return True

# Returns length of the longest substring
# with all distinct characters.


def longestUniqueSubsttr(str):

    n = len(str)

    # Result
    res = 0

    for i in range(n):
        for j in range(i, n):
            if (areDistinct(str, i, j)):
                res = max(res, j - i + 1)

    return res
```

```python
def lengthOfLongestSubstring(s):
        if len(set(s)) == len(s):                       # Checking if the string is unique
            return len(s)
        substring = ''
        maxLen = 1
        for i in s:
            if i not in substring:                      # If given character is not in substring we are generating, we add it to the substring
                substring = substring + i
                maxLen = max(maxLen, len(substring))    # We update the maxlength if our substring is longer than the current max.
            else:                                       # Else, if the character is already part of the substring, we need to break the flow
                substring = substring.split(i)[1] + i   # And create a new substring from that point onwards and this new substring is build from the next character. Process repeats till the loop ends.
        return maxLen
```
